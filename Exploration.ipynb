```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corpus= pd.read_json('Data\Archive.json')
cl = pd.read_json('Data\cluster_df.json')
```

```python
def returnLabels(cl_id):
    return cl[cl['cluster_id']==cl_id].cluster_label.unique()[0]
```

```python
cluster_info = pd.DataFrame(corpus.groupby('cluster_id')['DocId'].agg(num_of_docs = 'count')).reset_index()
cluster_info['cluster_name'] = cluster_info['cluster_id'].apply(returnLabels)
cluster_info.sort_values(by = 'num_of_docs', ascending = False, inplace = True)
```

```python
fig1 = sns.barplot(cluster_info[0:5],
            x="num_of_docs",
            y="cluster_name",
            palette = 'rocket')
```

```python
fig1 = sns.barplot(cluster_info[1:11],
            x="num_of_docs",
            y="cluster_name",
            palette = 'rocket')
```
